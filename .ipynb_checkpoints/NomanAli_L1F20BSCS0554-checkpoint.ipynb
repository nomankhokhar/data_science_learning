```python
class Book:
    def __init__(self, title, author, genre, price, quantity):
        self.title = title
        self.author = author
        self.genre = genre
        self.price = price
        self.quantity = quantity

class InventoryManagement:
    
    def __init__(self):
        self.inventory = {}

    def add_book(self, title, author, genre, price, quantity):
        new_book = Book(title, author, genre, price, quantity)
        self.inventory[title] = (new_book.author, new_book.genre, new_book.price, new_book.quantity)
        print(f"Book '{title}' added to the inventory.")

    def remove_book(self, title):
        if title in self.inventory:
            del self.inventory[title]
            print(f"Book '{title}' removed from the inventory.")
        else:
            print(f"Book '{title}' not found in the inventory.")

    def update_quantity(self, title, new_quantity):
        if title in self.inventory:
            author, genre, price, _ = self.inventory[title]
            self.inventory[title] = (author, genre, price, new_quantity)
            print(f"Quantity of '{title}' updated to {new_quantity}.")
        else:
            print(f"Book '{title}' not found in the inventory.")

    def display_all_books(self):
        print("All available books:")
        for title, (author, genre, price, quantity) in self.inventory.items():
            print(f"Title: {title}, Author: {author}, Genre: {genre}, Price: {price}, Quantity: {quantity}")

    def search_by_title_or_author(self, keyword):
        matching_books = [(title, info) for title, info in self.inventory.items() if keyword.lower() in title.lower() or keyword.lower() in info[0].lower()]
        if matching_books:
            print(f"Books matching '{keyword}':")
            for title, (author, genre, price, quantity) in matching_books:
                print(f"Title: {title}, Author: {author}, Genre: {genre}, Price: {price}, Quantity: {quantity}")
        else:
            print(f"No books found matching '{keyword}'.")

    def display_books_by_genre(self, genre):
        matching_books = [(title, info) for title, info in self.inventory.items() if genre.lower() == info[1].lower()]
        if matching_books:
            print(f"Books in the genre '{genre}':")
            for title, (author, genre, price, quantity) in matching_books:
                print(f"Title: {title}, Author: {author}, Genre: {genre}, Price: {price}, Quantity: {quantity}")
        else:
            print(f"No books found in the genre '{genre}'.")

inventory_manager = InventoryManagement()
inventory_manager.add_book("Kaluuuuu", "Noman Ali", "lolololololol", 199, -100)
inventory_manager.add_book("Kaluuuu", "Noman Ali", "lolololololol", 199, -100)
inventory_manager.add_book("Kaluuu", "Noman Ali", "lolololololol", 199, -100)
inventory_manager.display_all_books()

inventory_manager.search_by_title_or_author("Kaluuu")

inventory_manager.display_books_by_genre("lolololololol")

inventory_manager.update_quantity("Kaluuu", 25)

inventory_manager.remove_book("1984")

inventory_manager.display_all_books()
```

```python
def check_palindrome(string):
    value = str(string)
    for i in range(len(value)):
        if value[i] != value[len(value) -1 - i]:
            return False
    return True

def check_symmetrical(string):
    value = str(string)
    index = len(value)/2
    first_half = value[0:int(index)]
    second_half = value[int(index):]
    if first_half == second_half:
        return True
    return False


def main():
    user_set = eval(input("Enter a set: "))
    result_list = []
    for value in user_set:
        if check_palindrome(value):
            result_list.append((value, "palindrome"))
        elif check_symmetrical(value):
            result_list.append((value, "symmetrical"))

    print("List:", result_list)

if __name__ == "__main__":
    main()
```
